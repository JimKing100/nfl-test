```
# Imports
import pandas as pd
```

```
# Load the data
defense_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data-predictions/week1-pred-defense.csv')
offense_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data-predictions/week1-pred-offense-norookies.csv')
rookies_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data-predictions/week1-pred-offense-rookies.csv')
```

```
# Merge the three prediction files - defense, offense with no rookies and the rookies
predictions_df = pd.concat([defense_df, offense_df, rookies_df], ignore_index=True)
```

```
predictions_df.head()
```

```
predictions_df.tail()
```

```
predictions_df['rank-pred'] = predictions_df['week1-pred'].rank(method='min', ascending=False)
predictions_df['rank-act'] = predictions_df['week1-act'].rank(method='min', ascending=False)
```

```
predictions_df.head()
```

```
predictions_df = predictions_df.sort_values(by=['rank-pred'])
```

```
predictions_df.head(50)
```

```
# Save the results to .csv file
predictions_df.to_csv('/content/predictions-week1.csv', index=False)
```
