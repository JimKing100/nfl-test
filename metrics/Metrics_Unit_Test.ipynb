```
# Imports
import pandas as pd
import numpy as np

from sklearn.metrics import mean_absolute_error
```

```
# Load the raw data
w1_results_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data-predictions-nn/week1-pred-defense.csv')
```

```
#### The week 1 predictions

# week1-cur = 2018 total points
# week1-pred = predicted points for the season
# week1-act = actual points for the season

# weekn-cur = week (n-1) actual points
# weekn-pred = predicted points for the rest of the season (n-17)
# weekn-act = actual points for the rest of the season (n-17)

w1_results_df.head()
```

```
# Calculate the MAE for predicted points vs. actual points for NN
# Calculate the MAE for current points using the average of previous weeks
for i in range(1, 18):

    filename = 'https://raw.githubusercontent.com/JimKing100/nfl-test/master/data-predictions-nn/week' + str(i) + '-pred-defense.csv'
    
    # Column names
    week_cur = 'week' + str(i) + '-cur'
    week_pred = 'week' + str(i) + '-pred'
    week_act = 'week' + str(i) + '-act'

    # Weekly predictions
    results_df = pd.read_csv(filename)

    # Create the current points list using 2018 points in week 1 and average points going forward
    if i == 1:
        week_current = results_df['week1-cur'].tolist()
    else:
        # for each player (element) calculate the average points (element/(i-1)) and multiply by remaining games (17-(i-1))
        # the 17th week is 0 and represents the bye week (17 weeks and 16 games)
        week_list = results_df[week_cur].tolist()
        week_current = [((element / (i -1)) * (17 - (i -1))) for element in week_list]

    # Creat the prediction and actual lists
    week_pred = results_df[week_pred].tolist()
    week_act = results_df[week_act].tolist()

    # Calculate the MAE for predicted vs. actual
    week_pa_mae = mean_absolute_error(week_act, week_pred)
    print('MAE predicted vs actual week {0:2d} {1:3.2f}'.format(i, week_pa_mae))

    # Calculate the MAE for current vs. actual
    week_ca_mae = mean_absolute_error(week_act, week_current)
    print('MAE current vs actual week {0:2d} {1:3.2f}'.format(i, week_ca_mae), '\n')
```

```
# Calculate the MAE for predicted points vs. actual points for NN
# Calculate the MAE for predicted points vs. actual points for ARIMA
for i in range(1, 18):

    filename_nn = 'https://raw.githubusercontent.com/JimKing100/nfl-test/master/data-predictions-nn/week' + str(i) + '-pred-defense.csv'
    filename_ar = 'https://raw.githubusercontent.com/JimKing100/nfl-test/master/data-predictions-final/week' + str(i) + '-pred-defense.csv'
    
    # Column names
    week_pred = 'week' + str(i) + '-pred'
    week_act = 'week' + str(i) + '-act'

    # Weekly predictions
    results_nn_df = pd.read_csv(filename_nn)
    results_ar_df = pd.read_csv(filename_ar)

    # Creat the NN and ARIMA prediction lists
    week_pred_nn = results_nn_df[week_pred].tolist()
    week_pred_ar = results_ar_df[week_pred].tolist()
    week_act = results_nn_df[week_act].tolist()

    # Calculate the MAE for NN vs. actual
    week_nn_mae = mean_absolute_error(week_act, week_pred_nn)
    print('MAE NN predicted vs actual week {0:2d} {1:3.2f}'.format(i, week_nn_mae))

    # Calculate the MAE for ARIMA vs. actual
    week_ar_mae = mean_absolute_error(week_act, week_pred_ar)
    print('MAE ARIMA predicted vs actual week {0:2d} {1:3.2f}'.format(i, week_ar_mae), '\n')
```
