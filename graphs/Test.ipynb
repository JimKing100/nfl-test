```
mae_df = pd.read_csv('https://raw.githubusercontent.com/Lambda-School-Labs/nfl-fantasy-ds/master/data/metrics/metrics.csv')
mae_df.head()
```

```
column_names = ['Week', 'Model', 'MAE']
graph_df = pd.DataFrame(columns = column_names)

for i in range (1, 18):
    week = mae_df['week'].loc[(mae_df['week']==i)].iloc[0]
    model = 'NN'
    mae = mae_df['nn'].loc[(mae_df['week']==i)].iloc[0]
    graph_df = graph_df.append({'Week': week,
                                'Model': model,
                                'MAE': mae
    }, ignore_index=True)

    week = mae_df['week'].loc[(mae_df['week']==i)].iloc[0]
    model = 'ARIMA'
    mae = mae_df['arima'].loc[(mae_df['week']==i)].iloc[0]
    graph_df = graph_df.append({'Week': week,
                                'Model': model,
                                'MAE': mae
    }, ignore_index=True)

    week = mae_df['week'].loc[(mae_df['week']==i)].iloc[0]
    model = 'Average'
    mae = mae_df['average'].loc[(mae_df['week']==i)].iloc[0]
    graph_df = graph_df.append({'Week': week,
                                'Model': model,
                                'MAE': mae
    }, ignore_index=True)

    graph_df = graph_df.sort_values(by=['Model', 'Week'])

graph_df
```

```
fig = px.line(graph_df, x='Week', y='MAE', color='Model', render_mode='svg', title='Mean Average Error for Each Model')
fig.update_xaxes(nticks=17)
fig.update_layout(title={'x': 0.5, 'xanchor': 'center'})
fig.show()
```

```
correct_df = pd.read_csv('https://raw.githubusercontent.com/Lambda-School-Labs/nfl-fantasy-ds/master/data/metrics/ban_metrics.csv')
correct_df.head()
```

```
column_names = ['Week', 'Model', 'Percent Correct']
graph1_df = pd.DataFrame(columns = column_names)

for i in range (1, 18):
    week = correct_df['week'].loc[(correct_df['week']==i)].iloc[0]
    model = 'NN'
    correct = correct_df['nn-pct'].loc[(correct_df['week']==i)].iloc[0]
    graph1_df = graph1_df.append({'Week': week,
                                 'Model': model,
                                 'Percent Correct': correct
    }, ignore_index=True)

    week = correct_df['week'].loc[(correct_df['week']==i)].iloc[0]
    model = 'ARIMA'
    correct = correct_df['arima-pct'].loc[(correct_df['week']==i)].iloc[0]
    graph1_df = graph1_df.append({'Week': week,
                                 'Model': model,
                                 'Percent Correct': correct
    }, ignore_index=True)

    week = correct_df['week'].loc[(correct_df['week']==i)].iloc[0]
    model = 'Average'
    correct = correct_df['avg-pct'].loc[(correct_df['week']==i)].iloc[0]
    graph1_df = graph1_df.append({'Week': week,
                                 'Model': model,
                                 'Percent Correct': correct
    }, ignore_index=True)

    graph1_df = graph1_df.sort_values(by=['Model', 'Week'])

graph1_df
```

```
fig = px.line(graph1_df, x='Week', y='Percent Correct', color='Model', render_mode='svg', title='Percent Correct for Each Model')
fig.update_xaxes(nticks=17)
fig.update_layout(title={'x': 0.5, 'xanchor': 'center'})
fig.show()
```
