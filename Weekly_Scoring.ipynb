```
import pandas as pd
```

```
qb_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data/qb-2019-weekly.csv')
rec_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data/rec-2019-weekly.csv')
rush_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data/rush-2019-weekly.csv')
```

```
def scoring(p, w):
  # Passing TD
  if p in qb_df['Player'].values:
      score1 = qb_df['Pass TD'].loc[(qb_df['Player']==p) & (qb_df['Week']==w)]
      pass_td = score1.iloc[0]
  else:
      pass_td = 0

  # Receiving TD
  if p in rec_df['Player'].values:
      score2 = rec_df['Rec TD'].loc[(rec_df['Player']==p) & (rec_df['Week']==w)]
      rec_td = score2.iloc[0]
  else:
      rec_td = 0

  # Rushing TD
  if p in rush_df['Player'].values:
      score3 = rush_df['Rush TD'].loc[(rush_df['Player']==p) & (rush_df['Week']==w)]
      rush_td = score3.iloc[0]
  else:
      rush_td = 0
  
  # Passing Yards
  if p in qb_df['Player'].values:
      score4 = qb_df['Pass Yds'].loc[(qb_df['Player']==p) & (qb_df['Week']==w)]
      pass_yds = score4.iloc[0]
  else:
      pass_yds = 0

  # Receiving Yards
  if p in rec_df['Player'].values:
      score5 = rec_df['Rec Yds'].loc[(rec_df['Player']==p) & (rec_df['Week']==w)]
      rec_yds = score5.iloc[0]
  else:
      rec_yds = 0

  # Rushing Yards
  if p in rush_df['Player'].values:
      score6 = rush_df['Rush Yds'].loc[(rush_df['Player']==p) & (rush_df['Week']==w)]
      rush_yds = score6.iloc[0]
  else:
      rush_yds = 0

  # Interceptions
  if p in qb_df['Player'].values:
      score7 = qb_df['Pass Int'].loc[(qb_df['Player']==p) & (qb_df['Week']==w)]
      qb_int = score7.iloc[0]
  else:
      qb_int = 0
  
  score = (pass_td * 4) + (rec_td * 6) + (rush_td * 6) + (pass_yds / 25) + (rec_yds / 10) + (rush_yds / 10) - (qb_int *2)
  return score
```

```
player = 'Lamar Jackson'
week = 1

week1_score = scoring(player, week)
week1_score
```
