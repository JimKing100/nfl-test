```
qb_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data/qb-2019-weekly.csv')
rec_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data/rec-2019-weekly.csv')
rush_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data/rush-2019-weekly.csv')
rank_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data/teamranking-2019-weekly.csv')
```

```
def pull_stats(df, col, p, w, o):
    opp = o
    if (df[col].loc[(df['Player']==p) & (df['Week']==w)].empty):
        stat = 0
    else:
        stat = df[col].loc[(df['Player']==p) & (df['Week']==w)].iloc[0]
        opp = df['Opp'].loc[(df['Player']==p) & (df['Week']==w)].iloc[0]
    return stat, opp
```

```
def scoring(p, w):
    opponent = 'Bye'

    # Passing TD
    pass_td, opponent = pull_stats(qb_df, 'Pass TD', p, w, opponent)

    # Receiving TD
    rec_td, opponent = pull_stats(rec_df, 'Rec TD', p, w, opponent)

    # Rushing TD
    rush_td, opponent = pull_stats(rush_df, 'Rush TD', p, w, opponent)
  
    # Passing Yards
    pass_yds, opponent = pull_stats(qb_df, 'Pass Yds', p, w, opponent)

    # Receiving Yards
    rec_yds, opponent = pull_stats(rec_df, 'Rec Yds', p, w, opponent)

    # Rushing Yards
    rush_yds, opponent = pull_stats(rush_df, 'Rush Yds', p, w, opponent)

    # Interceptions
    qb_int, opponent = pull_stats(qb_df, 'Pass Int', p, w, opponent)

    score = (pass_td * 4) + (rec_td * 6) + (rush_td * 6) + (pass_yds / 25) + (rec_yds / 10) + (rush_yds / 10) - (qb_int *2)
    return score, opponent
```

```
def add_row(df, p, value_list):
    df = df.append({'Player': p,
                    'Week1': value_list[0], 
                    'Week2': value_list[1],
                    'Week3': value_list[2], 
                    'Week4': value_list[3], 
                    'Week5': value_list[4], 
                    'Week6': value_list[5], 
                    'Week7': value_list[6], 
                    'Week8': value_list[7], 
                    'Week9': value_list[8],
                    'Week10': value_list[9],
                    'Week11': value_list[10],
                    'Week12': value_list[11],
                    'Week13': value_list[12],
                    'Week14': value_list[13],
                    'Week15': value_list[14],
                    'Week16': value_list[15], 
                    'Week17': value_list[16] 
                     }, ignore_index=True)
    return df
```

```
def compare(p1, p2):

    column_names = ['Player', 'Week1', 'Week2', 'Week3', 'Week4', 'Week5', 'Week6', 'Week7', 'Week8', 'Week9',
                    'Week10', 'Week11', 'Week12', 'Week13', 'Week14', 'Week15', 'Week16', 'Week17']
    df1 = pd.DataFrame(columns = column_names)
    df2 = pd.DataFrame(columns = column_names)
    weekly_score1 = []
    weekly_score2 = []
    weekly_rank1 = []
    weekly_rank2 = []

    for i in range(1,18):
        score1, opp1 = scoring(p1, i)
        weekly_score1.append(score1)
        weekly_rank1.append(opp1)

    for i in range(1,18):
        score2, opp2 = scoring(p2, i)
        weekly_score2.append(score2)
        weekly_rank2.append(opp2)

    df1 = add_row(df1, p1, weekly_score1)
    df1 = add_row(df1, p2, weekly_score2)

    df2 = add_row(df2, p1, weekly_rank1)
    df2 = add_row(df2, p2, weekly_rank2)
  
    return df1, df2
```

```
player1 = 'Tom Brady'
player2 = 'Jimmy Garoppolo'

compare_df, ranking_df = compare(player1, player2)
```
