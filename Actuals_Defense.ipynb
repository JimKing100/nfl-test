```
# Pull the stat for the defense_db team points
def pull_stats(df1, df2, t, y, w):
    t_points = df1[y].loc[(df1['team']==t) & (df1['week']==w)].iloc[0]

    team_nameh = df2['h'].loc[(df2['seas']==int(y)) & (df2['wk']==w) & ((df2['v']==t) | (df2['h']==t))].iloc[0]
    team_namev = df2['v'].loc[(df2['seas']==int(y)) & (df2['wk']==w) & ((df2['v']==t) | (df2['h']==t))].iloc[0]

    if team_nameh == t:
        p_points = df2['ptsh'].loc[(df2['seas']==int(y)) & (df2['wk']==w) & ((df2['v']==t) | (df2['h']==t))].iloc[0]
    else:
        p_points = df2['ptsv'].loc[(df2['seas']==int(y)) & (df2['wk']==w) & ((df2['v']==t) | (df2['h']==t))].iloc[0]

    d_points = t_points + p_points
    
    return d_points
```

```
team = 'ARI'
year = '2000'
week = 1

weekly_dpoints = pull_stats(defense_df, game_df, team, year, week)
weekly_dpoints
```
