```
import pandas as pd
```

```
player_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data-revised/players.csv')
defense_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data-revised/defense.csv')
game_df = pd.read_csv('https://raw.githubusercontent.com/JimKing100/nfl-test/master/data-revised/game.csv')
```

```
# Pull the stats for the
def pull_stats(df, col, t, w):
    if (df[col].loc[(df['team']==t) & (df['week']==w)].empty):
        stat = 0
    else:
        stat = df[col].loc[(df['team']==t) & (df['week']==w)].iloc[0]
    return stat
```

```
def scoring(t, y, w):

    # Defensive team fantasy points
    f_points = pull_stats(defense_df, y, t, w)

    score = f_points
    return score
```

```
team = 'ARI'
year = '2000'
week = 1

weekly_score = scoring(team, year, week)
weekly_score
```
