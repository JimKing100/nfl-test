```
# Pull the stats to calculate the defense_db team points
def pull_stats(df1, df2, t, y, w, g):

    t_points = df1[y].loc[(df1['team']==t) & (df1['week']==g)].iloc[0]

    team_nameh = df2['h'].loc[(df2['seas']==int(y)) & (df2['wk']==w) & ((df2['v']==t) | (df2['h']==t))].iloc[0]
    team_namev = df2['v'].loc[(df2['seas']==int(y)) & (df2['wk']==w) & ((df2['v']==t) | (df2['h']==t))].iloc[0]

    if team_nameh == t:
        p_points = df2['ptsv'].loc[(df2['seas']==int(y)) & (df2['wk']==w) & ((df2['v']==t) | (df2['h']==t))].iloc[0]
    else:
        p_points = df2['ptsh'].loc[(df2['seas']==int(y)) & (df2['wk']==w) & ((df2['v']==t) | (df2['h']==t))].iloc[0]
    
    if p_points < 7:
        pa_points = 4
    if (p_points > 6) & (p_points < 14):
        pa_points = 3
    if (p_points > 13) & (p_points < 18):
        pa_points = 1
    if (p_points > 17) & (p_points < 28):
        pa_points = 0
    if (p_points > 27) & (p_points < 35): 
        pa_points = -1
    if (p_points > 34) & (p_points < 46):
        pa_points = -3
    if (p_points > 45):
        pa_points = -5

    d_points = t_points + pa_points

    return d_points
```
